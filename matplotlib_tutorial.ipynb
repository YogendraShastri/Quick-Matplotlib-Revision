## What is Matplotlib?
- **Matplotlib** is a powerful Python library used for data visualization. It helps you create static, animated, and interactive plots in Python.
- If you want to convert your boring data into interactive plots and graphs, **Matplotlib** is the tool for you.

## Why data visualization
- Sometimes when we look at raw data, it is very difficult for our minds to understand and draw conclusions from it. But when the same data is shown using visuals like charts or graphs, it becomes much easier to understand.
- Visuals help us quickly see patterns, trends, and insights that would be hard to find by just looking at a list of numbers. This is why data visualization is so important â€” it helps us find meaningful information and make better decisions.

## Bar Plot
- A bar plot, also known as a bar chart, is a visual representation of data using rectangular bars.
- The length or height of each bar corresponds to the value it represents.
- Bar plots are used to compare different categories or groups, displaying the relationship between a categorical variable.

**For Example**
- Ok now your company know over the year, company had a good growth, and ask you to present the products which truly contributes in this growth
- lets create bar chart between product and their sales.

```python
# Each product sales this year on monthly basis
Product_A = [0, 20, 25, 10, 15, 30, 35, 32, 30, 32, 33, 35]
Product_B = [0, 15, 20, 15, 10, 20, 25, 20, 23, 25, 27, 30]
Product_C = [0, 18, 22, 15, 20, 20, 18, 20, 22, 23, 25, 25]
```

```python
Product_A_TotalSales = sum(Product_A)
Product_B_TotalSales = sum(Product_B)
Product_C_TotalSales = sum(Product_C)

Total_sales = [Product_A_TotalSales, Product_B_TotalSales, Product_C_TotalSales]
Product_Names = ['Product_A', 'Product_B', 'Product_C']
plt.bar(Product_Names, Total_sales, color=['blue', 'green', 'yellow'])
plt.xlabel("Products")
plt.ylabel("Total Sales")
plt.title('Total Yearly Sales per Product')
plt.show()
```

```python
# Let's say you want to compare sales in June
Month = 'June'
Sales_June = [Product_A[5], Product_B[5], Product_C[5]]

plt.figure(figsize=(8, 6))
plt.bar(['Product A', 'Product B', 'Product C'], Sales_June, color=['skyblue', 'orange', 'lightgreen'])
plt.xlabel('Products')
plt.ylabel('Sales in June')
plt.title(f'Product Sales in {Month}')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

- above we see vertical bar graph 
- now will see horizontal bar graph, for that we just need to use barh instead of bar.

```python
Month = 'June'
Sales_June = [Product_A[5], Product_B[5], Product_C[5]]

plt.figure(figsize=(8, 6))
plt.barh(['Product A', 'Product B', 'Product C'], Sales_June, color=['skyblue', 'orange', 'lightgreen'])
plt.xlabel('Products')
plt.ylabel('Sales in June')
plt.title(f'Product Sales in {Month}')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**What is xticks in a Bar Graph?**

**xticks** controls the position and labeling of ticks (marks) on the x-axis.

It is used to:
- Customize the position of the ticks.
- Change the labels.
- Rotate labels for better visibility.

```python
import numpy as np
import matplotlib.pyplot as plt

labels = ['Q1', 'Q2', 'Q3', 'Q4']
product_A = [10, 15, 20, 25]
product_B = [12, 18, 24, 22]

x = np.arange(len(labels))  # [0, 1, 2, 3] for positions
width = 0.35  # Width of each bar

plt.bar(x - width/2, product_A, width, label='Product A', color='skyblue')
plt.bar(x + width/2, product_B, width, label='Product B', color='orange')

plt.xlabel('Quarters')
plt.ylabel('Sales')
plt.title('Quarterly Sales Comparison')

# Custom xticks
plt.xticks(ticks=x, labels=labels)

plt.legend()
plt.show()
```

## Histogram
- A histogram is a type of bar chart that shows how frequently different ranges of values appear in your data.
- It groups data into intervals called **bins**.
- Itâ€™s used to visualize the distribution of a dataset.

**Example :**
- lets continue the example
- Now the company wants to analyze how frequently different sales amounts occur for each product.
- But first lets see a example to learn how Histogram give us insides with a simple example.
  



```python
import numpy as np
# 40 students marks
students_marks = [np.random.randint(0,100) for i in range(0,40)]

# want to know how many students lies in each range
bins = 5
plt.hist(students_marks, bins, color='blue', edgecolor='black')
plt.xlabel("Numbers in Range")
plt.ylabel("Student Count")
plt.title("Student Mark Analysis")
plt.show()
```

**Lets understand now**
- From the above example, we can conclude a few things:
- Bins are interval ranges that group your continuous data.
- In the example, since we defined bins=5, matplotlib will try to fit the data such that, all data comes under 5 range bins or bars.
- if we edit the bins as 7, it will try to fit the data into 7 bins, you can see 7 bars.

```python
import matplotlib.pyplot as plt

#Now the company wants to analyze how frequently different sales amounts occur for each product.
Product_A = [0, 20, 25, 10, 15, 30, 35, 32, 30, 32, 33, 35]

plt.hist(Product_A, bins=7, color='skyblue', edgecolor='black')

plt.title('Product A Sales Frequency Distribution')
plt.xlabel('Sales Range')
plt.ylabel('Number of Months')
plt.show()
```

**Understanding**
- From the above plot, we can understand that:
- The sales data for Product A is grouped into intervals (bins).
- Since we used bins=7 for the sales data ranging from 0 to 35, Matplotlib automatically divided the total range into 7 equal intervals.
- Each bin covers a range of (maximum - minimum) Ã· number of bins = (35 - 0) Ã· 7 = 5 units.
- So, the bins are: 0-5, 5-10, 10-15, 15-20, 20-25, 25-30, 30-35
- According to the plot:
- The range 30-35 contains 7 data points: [30, 35, 32, 30, 32, 33, 35], rest 5 are 1-1 unit.
- Total number of data points is 5+7 = 12.

## Scatter plot
- A scatter plot is used to show the relationship (correlation) between two numerical variables.
- It helps us see: {Trends, Patterns, Outliers, Correlations (positive, negative, or none)}

**Example**
- The company now wants to know:
- "Does the amount spent on advertising each month correlate with monthly sales growth?"

```python
import matplotlib.pyplot as plt

# Data
Months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
Ad_Spend = [1000, 2000, 2500, 1500, 1800, 3000, 3500, 3200, 3300, 3700, 4000, 4200]
Total_Sales = [0, 53, 67, 40, 45, 70, 78, 72, 75, 80, 85, 90]

plt.figure(figsize=(10, 6))
plt.scatter(Ad_Spend, Total_Sales, color='purple', s=100, edgecolors='black')

plt.title('Advertising Spend vs. Monthly Sales')
plt.xlabel('Advertising Spend')
plt.ylabel('Total Sales (Units)')
plt.grid(True)
```

**Parameters**
- you can change the all the dot color and size all together
- also you can set visibility/transparancy using alpha=(0-1) parameter.
- we can change the marker as well.
- later will see cmaps.

```python
import matplotlib.pyplot as plt

# Data
Months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
Ad_Spend = [1000, 2000, 2500, 1500, 1800, 3000, 3500, 3200, 3300, 3700, 4000, 4200]
Total_Sales = [0, 53, 67, 40, 45, 70, 78, 72, 75, 80, 85, 90]

plt.figure(figsize=(10, 6))
c = ['r', 'g', 'b', 'c', 'y', 'r', 'g', 'b', 'c', 'y', 'r', 'y']
# Custom sizes
sizes = [100, 60, 70, 120, 400, 100, 60, 70, 120, 40, 50, 200]
plt.scatter(Ad_Spend, Total_Sales, color=c, s=sizes, edgecolors='black', alpha=0.5, marker="*")

plt.title('Advertising Spend vs. Monthly Sales')
plt.xlabel('Advertising Spend')
plt.ylabel('Total Sales (Units)')
plt.grid(True)
plt.show()
```

**What is a Colormap (cmap)?**
- A colormap is a range of colors that Matplotlib can automatically apply to your data based on a third variable.
- It maps numerical data to color gradients.
- cmaps color gradients names : https://matplotlib.org/stable/users/explain/colors/colormaps.html

```python
# example
import matplotlib.pyplot as plt
import numpy as np

Months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
Ad_Spend = [1000, 2000, 2500, 1500, 1800, 3000, 3500, 3200, 3300, 3700, 4000, 4200]
Total_Sales = [0, 53, 67, 40, 45, 70, 78, 72, 75, 80, 85, 90]

plt.figure(figsize=(10, 6))

scatter = plt.scatter(Ad_Spend, Total_Sales, c=Total_Sales, cmap='viridis', s=100, edgecolors='black')

plt.title('Advertising Spend vs. Monthly Sales (Colormap)')
plt.xlabel('Advertising Spend')
plt.ylabel('Total Sales (Units)')
plt.grid(True)

# Add colorbar
plt.colorbar(scatter, label='Sales (Units)')

plt.show()
```

- c= Total_sales tell, assign color according to this value,
- cmaps='viridis' tells which cmap to use
- colorbar represents the color range

## Pie Chart
- A pie chart shows parts of a whole as slices of a circle.
- Itâ€™s used to display percentage or proportional data.

```python
import matplotlib.pyplot as plt

sales = [300, 400, 100, 500]
products = ['A','B','C','D']

plt.pie(sales, labels=products)
plt.show()
```

**Key Parameters:**
- labels : Names of slices
- autopct : ("%0.1f%" tells till 1 floating point) Display percentage on slices
- startangle : Rotation angle of start position (rotate the pie chart)
- colors : Custom colors for slices
- explode : Pull slices out for emphasis
- shadow : Adds shadow for 3D effect (Boolean parameter)
- radius : Define the radius of the circle

```python
import matplotlib.pyplot as plt

sales = [300, 400, 100, 500]
products = ['A','B','C','D']
c = ['r','g','b','y'] # red green blue yellow
explode = [0, 0.1, 0, 0] # exploding green part
plt.pie(sales, labels=products, autopct="%0.1f%%", startangle=50, colors=c, explode=explode,shadow=True, radius=1.5)
plt.show()
```
